```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
```

```python
lrs = sns.load_dataset('anscombe')
lrs.head()
```

```python
X = lrs['x'].values.reshape(-1, 1)
y = lrs['y'].values
```

```python
pf = PolynomialFeatures(degree=2)
X_poly = pf.fit_transform(X)
```

```python
lr = LinearRegression()
lr.fit(X_poly, y)
```

```python
X_new = np.linspace(0, 52, 100).reshape(-1, 1)
X_new_poly = pf.transform(X_new)
```

```python
y_pred = lr.predict(X_new_poly)
```

```python
plt.scatter(X, y, label='Original Data', alpha=0.7)
plt.plot(X_new, y_pred, 'r-', label='Prediction', linewidth=2)
plt.title('Polynomial Regression on anscombe Dataset')
plt.xlabel('x')
plt.ylabel('y')

plt.tight_layout()
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()
```
